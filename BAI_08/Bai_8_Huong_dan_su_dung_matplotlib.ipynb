```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 2. Vẽ Biểu Đồ Đường (Line Chart)

# Dữ liệu mẫu
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Vẽ đồ thị
plt.plot(x, y, label='Dãy số nguyên tố', color='green', marker='o')

# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ đường')
plt.xlabel('Giá trị X')
plt.ylabel('Giá trị Y')

# Hiển thị chú thích
plt.legend()

# Hiển thị đồ thị
plt.show()
```

```python
# 3. Vẽ Biểu Đồ Cột (Bar Chart) 
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 5]

plt.bar(categories, values, color='skyblue')
plt.title('Biểu đồ cột đơn giản')
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
# 4. Vẽ Biểu Đồ Tròn (Pie Chart) 
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 30, 15, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Tỷ lệ sử dụng ngôn ngữ lập trình')
plt.axis('equal')  # Giữ biểu đồ tròn đúng hình tròn
plt.show()
```

```python
# 5. Vẽ Nhiều Biểu Đồ Trên Cùng Một Hình (Subplots) 
x = [1, 2, 3, 4, 5]
y1 = [i**2 for i in x]
y2 = [i**0.5 for i in x]

plt.subplot(1, 2, 1)  # 1 hàng, 2 cột, vẽ ở vị trí 1
plt.plot(x, y1)
plt.title('Bình phương')

plt.subplot(1, 2, 2)  # Vị trí 2
plt.plot(x, y2)
plt.title('Căn bậc hai')

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i**2 for i in x]       # Bình phương
y2 = [i**0.5 for i in x]     # Căn bậc hai
y3 = [i*2 for i in x]        # Nhân đôi
y4 = [i+3 for i in x]        # Cộng 3

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Bình phương')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Căn bậc hai')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Nhân đôi')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Cộng 3')

plt.tight_layout()
plt.show()
```

```python
# 6. Tùy Chỉnh Thêm 
# Có thể thêm lưới, thay đổi kích thước hình vẽ, định dạng font, lưu đồ thị:
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='x^2', linestyle='--', color='red')
plt.grid(True)
plt.legend()
plt.savefig('do_thi.png')  # Lưu hình
plt.show()
```

```python
# 7. Biểu đồ phân tán (Scatter Plot)

# Tạo dữ liệu ngẫu nhiên
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(10, 300, 50)

# Vẽ biểu đồ phân tán
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar()  # Hiển thị thanh màu
plt.title('Biểu đồ phân tán')
plt.show()
```

```python
# 8.Biểu đồ dạng hộp (Box Plot) 
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# patch_artist Khi đặt là True, Matplotlib sẽ tô màu phần hộp (box) bằng màu mặc định hoặc màu do bạn chọn (thay vì chỉ có viền)
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Biểu đồ hộp')
plt.xticks([1, 2, 3], ['Nhóm 1', 'Nhóm 2', 'Nhóm 3'])
plt.show()
```

```python
# 9.Biểu đồ histogram (Phân phối tần suất)
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram - Phân phối chuẩn')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.show()
```

```python
# Tạo ra một ma trận ngẫu nhiên 10 hàng × 12 cột, với mỗi phần tử là số thực trong khoảng [0, 1)
np.random.rand(10, 12)
```

```python
# 10. Vẽ heatmap với thư viện seaborn (kết hợp matplotlib) 
data = np.random.rand(10, 12)

sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap dữ liệu ngẫu nhiên')
plt.show()
```

```python
# 11. Vẽ từ DataFrame của Pandas 
# Tạo DataFrame giả
data = {
    'Tháng': ['1', '2', '3', '4', '5'],
    'Doanh thu': [100, 120, 90, 150, 170],
    'Chi phí': [50, 60, 55, 80, 90]
}
df = pd.DataFrame(data)
df
```

```python
# Vẽ biểu đồ cột nhóm
df.plot(x='Tháng', kind='bar', figsize=(8,5))
plt.title('Doanh thu và Chi phí theo tháng')
plt.ylabel('Triệu đồng')
plt.grid(True)
plt.show()
```

```python
# 12. 3D Plot với matplotlib
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z, c='r', marker='^')
ax.set_title('Biểu đồ phân tán 3D')
plt.show()
```
