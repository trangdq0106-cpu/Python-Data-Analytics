## 1. Giới thiệu về Python 

Python là một ngôn ngữ lập trình nguồn mở phổ biến, cấp cao (Hight level), đơn giản (simple), hướng đối tượng để phát triển phần mềm và kịch bản chung, phát triển web và khoa học dữ liệu, một số tính năng quan trọng khác của ngôn ngữ là:  
 - **Ngôn ngữ lập trình cấp cao (Hight level)**: Thân thiện với ngôn ngữ tự nhiên, Dễ hiểu, Dễ để debug tìm lỗi, Dễ dàng để bảo trì nâng cấp, Có thể chạy ở nhiều môi trường, Có thể chạy trên bất cứ nền tảng nào  
 - **Ngôn ngữ thông dịch (Interpreted Language)** : Vì mã nguồn của nó không được biên dịch trực tiếp thành mã máy (machine code) trước khi thực thi, mà thay vào đó, nó được thực thi từng dòng một bởi một chương trình gọi là trình thông dịch (interpreter). Ngoài ra, so với các ngôn ngữ phổ biến khác như C, C ++, Java, v.v., mã Python không yêu cầu biên dịch, giúp gỡ lỗi dễ dàng hơn.  
Quá trình biên dịch nội bộ:  
Python sử dụng một bước trung gian gọi là biên dịch sang bytecode. 
Khi bạn chạy một chương trình Python, trình thông dịch sẽ biên dịch mã nguồn thành mã bytecode (.pyc hoặc .pyo trong thư mục __pycache__).  
Bytecode này là một dạng mã trung gian, không phải mã máy, giúp tăng tốc độ thực thi trong lần chạy tiếp theo.  
Bytecode sau đó được thực thi bởi Python Virtual Machine (PVM).  

 - **Ngôn ngữ đa nền tảng (Cross-platform Language)**: Python tương thích với nhiều hệ điều hành/ nền tảng, bao gồm Windows, MacOS, Linux, v.v. Tính năng này cho phép mã Python có thể chạy trên các nền tảng di động, cho phép các lập trình viên viết mã một lần và chạy không thể phân biệt trên một số nền tảng.  
 - **Mã nguồn miễn phí và mở (Free and Open Source)**: Ngôn ngữ miễn phí để tải xuống và sử dụng. Người dùng có thể tải xuống Python từ trang web chính thức. Thêm vào đó, công chúng cũng có thể thấy mã nguồn vì Python là nguồn mở.  
 - **Hướng đối tượng (Object Oriented)**: Python có các tính năng lập trình hướng đối tượng, cho phép tạo và sử dụng các lớp, đối tượng và kế thừa. Nó khuyến khích mã mô -đun và có thể tái sử dụng, giúp phát triển các ứng dụng phức tạp dễ dàng hơn.
     
        - Lập trình hướng đối tượng (OOP) là kỹ thuật lập trình dựa trên “công nghệ đối tượng”, tạo ra các đối tượng trong code trừu tượng hóa các đối tượng thực tế trong thế giới thực. Đối tượng trong OOP có thuộc tính và phương thức. Chúng có thể tương tác qua lại lẫn nhau.

        - Lập trình hướng đối tượng OOP có 4 tính chất:
            *Encapsulation – tính đóng gói: các phương thức và dữ liệu có mối quan hệ với nhau được lưu vào một lớp để thuận tiện cho việc quản lý, sử dụng. Chỉ có phương thức nội tại của chính đối tượng mới có thể thay đổi trạng thái nội tại của nó.*
            *Abstraction – tính trừu tượng: chỉ tập trung vào những thuộc tính và phương thức cần thiết cho việc giải quyết vấn đề trong lập trình và bỏ qua các thông tin không quan trọng.* 
            *Inheritance – tính kế thừa: các đối tượng “con” có thể thừa hưởng các đặc tính có sẵn từ đối tượng “cha” mà không cần định nghĩa lại (tùy theo ngôn ngữ lập trình).*
            *Polymorphism – tính đa hình: các đối tượng không cùng một lớp, khi tiếp nhận cùng một thông điệp thì sẽ phản hồi theo những cách khác nhau.*
    
        - Lập trình hướng cấu trúc (POP) (Pascal, C, Foxpro): là kỹ thuật lập trình chia nhỏ một chương trình lớn thành các chương trình con (còn được gọi là các hàm). Mỗi hàm sẽ đảm nhiệm một chức năng khác nhau trong hệ thống. Quá trình phân nhỏ sẽ được thực hiện cho đến khi ra được các hàm đơn giản nhất. Mục đích của việc này là để đơn giản hóa cấu trúc của chương trình, thuận tiện cho việc kiểm tra, sửa đổi và thực thi một cách hiệu quả. 
        *Lập trình hướng cấu trúc POP có đặc điểm:*
            *Chỉ tập trung vào việc phát triển các hàm, ít chú trọng đến dữ liệu*
            *Dữ liệu của hệ thống di chuyển từ hàm này qua hàm khác, được dùng chung giữa các hàm.* 
            *Tuân theo hình thức tiếp cận top-down khi thiết kế chương trình* 
            *Dùng con trỏ hoặc biến toàn cục để liên kết các hàm với nhau*
    
        - So sánh lập trình hướng đối tượng và lập trình hướng cấu trúc
        Có rất nhiều điểm khác biệt giữa lập trình hướng đối tượng và lập trình hướng cấu trúc mà bạn nên quan tâm như: 
        Trọng tâm
            OOP chú ý vào dữ liệu hơn là thuật toán. POP chú ý đến việc xây dựng các hàm và thuật toán hơn là dữ liệu.  
            Sự phân chia
            OOP chia nhỏ chương trình thành các đối tượng. POP chia nhỏ chương trình thành các hàm con. 
            Chế độ truy cập
            Các từ khóa phạm vi truy cập trong OOP được chia thành Public, Private, Protected và Default. POP không có thành phần này. 
            Hướng tiếp cận khi thiết kế chương trình
            OOP tiếp cận từ dưới lên. POP tiếp cận từ trên xuống.
            Quá trình thực thi
            OOP cho phép các chức năng chạy cùng một lúc. POP cho phép các hàm và chức năng chạy lần lượt. 
            Truy cập dữ liệu
            OOP hạn chế truy cập dữ liệu giữa các đối tượng. POP cho phép dữ liệu tự do di chuyển trong hệ thống và các hàm có thể chia sẻ dữ liệu cho nhau. 
            Bảo mật
            OOP ẩn dữ liệu trong chế độ Public, Protected và Private nên có bảo mật cao. POP không có chế độ ẩn dữ liệu, độ an toàn thấp. 
            Thêm mới dữ liệu
            Hoạt động này có thể được thực hiện dễ dàng với các đối tượng trong OOP, còn POP thì khó hơn. 
            Nạp chồng/Đa hình
            OOP hỗ trợ nạp chồng các hàm, hàm tạo và toán tử còn POP thì không. 
            Ứng dụng
            OOP có thể áp dụng trong xây dựng các chương trình có độ phức tạp cao. POP chỉ nên được dùng với chương trình đơn giản.* 



 - **Thư viện tiêu chuẩn rộng lớn (Large Standard Library)**: Python chứa một thư viện tiêu chuẩn khổng lồ với nhiều lựa chọn các mô -đun và chức năng, vì vậy bạn không cần phải viết bất kỳ mã nào của riêng mình. Biểu thức tiêu chuẩn (Regular expressions), kiểm tra đơn vị (unit testing), trình duyệt web và các công cụ khác đều là một phần của thư viện khá lớn của Python.

 - **Hỗ trợ GUI (GUI Support)**: Có thể tạo/ phát triển các giao diện người dùng đồ họa trong Python bằng cách sử dụng các gói như PYQT5, PYQT4, WXPYTHON hoặc TK,KIWY.....

 - **Dễ sử dụng và bảo trì (expressive language)**: Python khá dễ dàng để chọn so với các ngôn ngữ khác như C, C'#' , JavaScript, v.v. Bất cứ ai cũng có thể chọn những điều cơ bản của ngôn ngữ lập trình tương đối dễ dàng trong vài giờ hoặc ngày. Python là một trong những ngôn ngữ lập trình thân thiện với người dùng nhất hiện có. Expressive Language (ngôn ngữ biểu đạt) trong lập trình là khả năng biểu diễn các ý tưởng một cách ngắn gọn, dễ đọc, dễ hiểu mà vẫn giữ được độ chính xác và hiệu quả. Python được coi là một ngôn ngữ biểu đạt cao nhờ vào cú pháp đơn giản và các tính năng mạnh mẽ giúp lập trình viên viết mã rõ ràng và hiệu quả hơn.     

 - **Mạnh mẽ**: Sự mạnh mẽ của Python được thể hiện các khả năng quản lý lỗi đáng gờm của nó, cho phép các lập trình viên tìm thấy và sửa lỗi trong mã của họ một cách hiệu quả. Những tính năng này bao gồm xử lý ngoại lệ tích hợp và thông tin theo dõi.

 - **Đổi kiểu biến tùy biến (Dynamically Typed)**: Python là ngôn ngữ được đổi kiểu biến tùy biến, tức là, loại biến được quyết định tại thời điểm chạy. Tính năng này giúp loại bỏ yêu cầu buột người dùng để khai báo loại biến trước.
    
Những tính năng này góp phần làm cho Python trở thành lựa chọn phổ biến cho người mới bắt đầu, các nhà phát triển có kinh nghiệm và các nhà khoa học. Do đó làm cho nó trở thành một ngôn ngữ đa năng cho một loạt các ứng dụng.

## 3.Định danh (Identifier)

Trong Python, định danh (identifier) là tên được sử dụng để nhận diện các thực thể như biến, hàm, lớp, mô-đun hoặc đối tượng.

## 6.Toán tử (Operator)

### Toán tử số học (Arithmetic Operators)  
Toán tử số học trong Python được sử dụng để thực hiện các phép toán trên số (int, float, complex). Python hỗ trợ các toán tử số học cơ bản như cộng, trừ, nhân, chia, chia lấy dư, lũy thừa, chia lấy phần nguyên.
| Toán tử | Ký hiệu | Mô tả | Ví dụ |
|---------|--------|-------|------|
| Cộng | `+` | Cộng hai số | `5 + 3 = 8` |
| Trừ | `-` | Trừ hai số | `5 - 3 = 2` |
| Nhân | `*` | Nhân hai số | `5 * 3 = 15` |
| Chia | `/` | Chia hai số (kết quả là `float`) | `5 / 2 = 2.5` |
| Chia lấy phần nguyên | `//` | Chia và lấy phần nguyên | `5 // 2 = 2` |
| Chia lấy dư | `%` | Lấy phần dư của phép chia | `5 % 2 = 1` |

### Toán tử so sánh (Comparison Operators)

Toán tử so sánh được sử dụng để so sánh hai giá trị. Kết quả của biểu thức so sánh luôn là một giá trị Boolean (True hoặc False).  
Bob có 2 cây kẹo và Marry cũng có 2 cây kẹo, nên số keo của Bob bằng với số kẹo của Marry  
| Toán tử | Ký hiệu | Mô tả | Ví dụ |
|---------|--------|-------|------|
| Bằng nhau | `==` | Trả về `True` nếu hai giá trị bằng nhau | `5 == 5` → `True` |
| Khác nhau | `!=` | Trả về `True` nếu hai giá trị không bằng nhau | `5 != 3` → `True` |
| Lớn hơn | `>` | Trả về `True` nếu vế trái lớn hơn vế phải | `5 > 3` → `True` |
| Nhỏ hơn | `<` | Trả về `True` nếu vế trái nhỏ hơn vế phải | `5 < 3` → `False` |
| Lớn hơn hoặc bằng | `>=` | Trả về `True` nếu vế trái lớn hơn hoặc bằng vế phải | `5 >= 5` → `True` |
| Nhỏ hơn hoặc bằng | `<=` | Trả về `True` nếu vế trái nhỏ hơn hoặc bằng vế phải | `3 <= 5` → `True` |  

✔ Toán tử so sánh giúp kiểm tra điều kiện trong Python.  
✔ Kết quả luôn là True hoặc False.  
✔ Có thể sử dụng cho số, chuỗi, danh sách, tuple, set,...  
✔ Cần cẩn thận khi so sánh kiểu dữ liệu khác nhau.  

    Chú ý : Toán tử so sánh bằng là == , còn phép gán bằng là =  
    Ví dụ : Gán biến a = 10 và b = 5 ; Toán tử so sánh a == b  trả về False.

```python
# 📌 Toán tử so sánh với chuỗi (String)
# Python cũng hỗ trợ toán tử so sánh với chuỗi. So sánh chuỗi dựa trên thứ tự từ điển (lexicographical order).
print("apple" == "banana")   # False
print("apple" < "banana")    # True (vì "a" đứng trước "b" trong bảng chữ cái)
print("car" > "bike")        # True (vì "c" đứng sau "b")
```

```python
# 📌 Toán tử so sánh với danh sách (List), Tuple và Set
# Các kiểu dữ liệu có thể so sánh dựa trên phần tử đầu tiên (và tiếp tục nếu bằng nhau).
print([1, 2, 3] == [1, 2, 3])  # True
print((1, 2, 3) < (1, 3, 2))   # True (so sánh từng phần tử)
print({1, 2, 3} == {3, 2, 1})  # True (Set không quan trọng thứ tự)
```

## Toán tử logic (Logical Operators)  

Toán tử logic được sử dụng để kết hợp hoặc kiểm tra nhiều điều kiện Boolean (True hoặc False).  
📌 Danh sách các toán tử logic trong Python  
| Toán tử | Ký hiệu | Mô tả | Ví dụ |
|---------|--------|-------|------|
| AND | `and` | Trả về `True` nếu cả hai điều kiện đều đúng | `(5 > 3) and (10 > 5) → True` |
| OR | `or` | Trả về `True` nếu ít nhất một điều kiện đúng | `(5 > 3) or (10 < 5) → True` |
| NOT | `not` | Đảo ngược giá trị Boolean | `not (5 > 3) → False` |  

📌 Toán tử logic với số và chuỗi  
Python không chỉ áp dụng toán tử logic cho True và False, mà còn có quy tắc sau:  

Giá trị "falsy": 0, "" (chuỗi rỗng), None, [] (danh sách rỗng)...  == > False  
Giá trị "truthy": Số khác 0, chuỗi không rỗng, danh sách có phần tử...   ==> True  

print(5 and 10)  # 10 (trả về giá trị cuối cùng nếu cả hai là True)  
print(0 and 10)  # 0 (trả về giá trị đầu tiên nếu nó là False)  

print("" or "Hello")  # "Hello" (vì chuỗi rỗng là False, nên lấy giá trị còn lại)  
print("Python" or "Hello")  # "Python" (vì giá trị đầu tiên là True, nên giữ nguyên)

### Toán tử thành phần (Membership Operators) trong Python  
Toán tử thành phần được sử dụng để kiểm tra xem một giá trị có tồn tại trong một chuỗi, list, tuple, set hoặc dictionary hay không. Python cung cấp hai toán tử thành phần:  
| Toán tử  | Mô tả | Ví dụ |
|----------|-------|------|
| `in`     | Trả về `True` nếu giá trị nằm trong chuỗi, danh sách, tập hợp hoặc từ điển | `"a" in "apple" → True` |

📌 Cú pháp  
input(prompt)

#### Hàm eval() trong Python  
Hàm eval() trong Python cho phép thực thi một biểu thức Python được truyền dưới dạng chuỗi và trả về kết quả của biểu thức đó. Nó đặc biệt hữu ích khi cần chuyển đổi đầu vào của input() thành kiểu dữ liệu phù hợp một cách linh hoạt.  
📌 Cú pháp  
eval(expression, globals=None, locals=None)  
- expression (bắt buộc): Chuỗi biểu thức Python cần thực thi.  
- globals (tùy chọn): Từ điển chứa các biến toàn cục có thể dùng trong biểu thức.  
- locals (tùy chọn): Từ điển chứa các biến cục bộ có thể dùng trong biểu thức.  
🚨 Lưu ý quan trọng: eval() có thể gây nguy hiểm nếu thực thi mã không đáng tin cậy, vì nó có thể thực hiện bất kỳ lệnh Python nào.

## Xuất dữ liệu: Output
#### Hàm print() trong Python  
Hàm print() trong Python được sử dụng để in dữ liệu ra màn hình. Đây là một trong những hàm được sử dụng nhiều nhất để hiển thị thông tin trong các chương trình Python.  
📌 Cú pháp  
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)  
📌 Các tham số:  
- *objects: Một hoặc nhiều đối tượng cần in ra.  
- sep (tùy chọn): Chuỗi phân cách giữa các đối tượng (mặc định là dấu cách ' ').  
- end (tùy chọn): Chuỗi kết thúc dòng (mặc định là '\n', tức xuống dòng).  
- file (tùy chọn): Đối tượng tệp (mặc định là sys.stdout, tức in ra màn hình).  
- flush (tùy chọn): Nếu True, dữ liệu được đẩy ngay ra màn hình mà không cần đợi bộ đệm.  

✅ Hiển thị kết quả của chương trình.  
✅ Ghi log và debug mã nguồn.  
✅ Xuất dữ liệu ra tệp hoặc màn hình giao diện dòng lệnh.
